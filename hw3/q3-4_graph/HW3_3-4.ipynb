```python
# Extracting data using regular expressions
import re

# Now we will read the content of the '3 error output.txt' and '4 error output.txt' files
with open('./3 error output.txt', 'r') as file:
    error_output_3 = file.read()
with open('./4 error output.txt', 'r') as file:
    error_output_4 = file.read()

# Regular expression matching size and unround
pattern = r"Average estimation error for size (\d+): unround:(\d+\.\d+)"

matches_3 = re.findall(pattern, error_output_3)
matches_4 = re.findall(pattern, error_output_4)

# Convert extracted data to the desired format
data_3_error = {
    "size": [int(size) for size, _ in matches_3],
    "unround": [float(unround) for _, unround in matches_3]
}

data_4_error = {
    "size": [int(size) for size, _ in matches_4],
    "unround": [float(unround) for _, unround in matches_4]
}
```
